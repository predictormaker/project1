```python
merged_df.columns = ['일자', 'CJ_ENM', 'IHQ', '덱스터', '쇼박스', '위지윅스튜디오', '콘텐트리중앙', 'CJ_CGV', '롯데쇼핑', '아센디오', 'NEW', '바른손이엔에이', '판타지오', '애니플러스', '스튜디오산타클로스']
merged_df.set_index('일자', inplace=True)
merged_df.reset_index(inplace=True)
merdf_num = merged_df.T.drop(index = '일자')
merdf_num
```

```python
# 시가총액 컬럼 생성
merged_df['시가총액'] = merdf_num.sum()
# 시가 총액 방식을 이용한 주가지수 컬럼 생성
merged_df['주가지수'] = merged_df['시가총액'] / 688191757200.0 * 100
merged_df.tail()
```

```python
# 수치 조정 전 그래프
import matplotlib.pyplot as plt
plt.figure()
merged_df.sort_values('일자').plot('일자', '주가지수')
plt.show()
```

```python
mer_int = merged_df.drop(columns='일자')
#  기업이 상장한 날의 기업개수 컬럼 생성
merged_df['기업개수'] = 14 - mer_int.isna().sum(1)
# 주가지수를 기업개수로 나누어 수정주가지수 생성(주가지수의 왜곡된 수치 수정) 
merged_df['수정주가지수'] = merged_df['주가지수'] / merged_df['기업개수'] * 4
# 수치 조정 후 그래프
plt.figure()
merged_df.sort_values('일자').plot('일자', '수정주가지수')
plt.show()
```
